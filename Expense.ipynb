```python
from faker import Faker
import random
import pandas as pd
fake = Faker()
categories = ["Food", "Transportation", "Savings","Bills", "Entertainment", "Shopping", "Healthcare","Rental"]
payment_modes = ["Cash", "Credit Card", "Debit Card", "UPI", "Net Banking"]

def generate_expense_data(num_entries=500):
    data = []
    for _ in range(num_entries):
        date = fake.date_between(start_date="-1y", end_date="today")  
        category = random.choice(categories)
        payment_mode = random.choice(payment_modes)
        amount_paid = round(random.uniform(10, 500), 2)  
        cashback = round(amount_paid * random.choice([0, 0.05, 0.1]), 2)  
        data.append([date, category, payment_mode, fake.sentence(), amount_paid, cashback])

    return pd.DataFrame(data, columns=["Date", "Category", "Payment Mode", "Description", "Amount Paid", "Cashback"])
df = generate_expense_data()
df.to_csv("expenses.csv", index=False)  
df.head()
```

```python
import streamlit as st
import pandas as pd

df = pd.read_csv("expenses.csv")

st.title("Personal Expense Tracker ðŸ“Š")

st.subheader("Total Spending by Category")
category_expense = df.groupby("Category")["Amount Paid"].sum().reset_index()
st.bar_chart(category_expense.set_index("Category"))

st.subheader("Total Spending by Payment Mode")
payment_expense = df.groupby("Payment Mode")["Amount Paid"].sum().reset_index()
st.bar_chart(payment_expense.set_index("Payment Mode"))

st.subheader("Total Cashback Earned")
st.write(f"ðŸ’° Total Cashback Received: **${df['Cashback'].sum():.2f}**")

if st.checkbox("Show Raw Data"):
    st.dataframe(df)
```
